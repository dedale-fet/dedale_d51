```python
%matplotlib inline
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
filter_folder = "/share/splinter/moraes/Dedale_Project/2016-06-13_simulated_catalogs/updated_filters/"
euclid_filters = os.path.join(filter_folder, "Euclid", "filters_galform")
lsst_filters = os.path.join(filter_folder, "LSST")
des_filters = os.path.join(filter_folder, "DES")
panstarrs_filters = os.path.join(filter_folder, "PanSTARRS")
```

```python
fig, axtup = plt.subplots(nrows=2, ncols=2, figsize=(13, 9))

for ft, c in zip("ugrizy", ["magenta", "blue", "green", "red", "orange", "black"]):
    ax = axtup[0][0]
    ax.set_title("LSST")
    lsst0 = np.loadtxt(os.path.join(lsst_filters, "lsst_total_%s.pb" % ft))
    ax.plot(lsst0[:, 0], lsst0[:, 1], color=c)
    ax.set_xlim(3000, 11500)
    ax.set_ylim(0, 1)
    ax.set_xlabel("$\lambda$ ($\AA$)", fontsize=16)
    ax.set_ylabel("throughput", fontsize=14)
    
    ax = axtup[0][1]
    ax.set_title("DES")
    des0 = np.loadtxt(os.path.join(des_filters, "des_%s.pb" % ft))
    ax.plot(des0[:, 0], des0[:, 1], color=c)
    ax.set_xlim(3000, 11500)
    ax.set_ylim(0, 1)
    ax.set_xlabel("$\lambda$ ($\AA$)", fontsize=16)
    ax.set_ylabel("throughput", fontsize=14)
    
    ax = axtup[1][0]
    ax.set_title("panSTARRS")
    if ft == "u":
        continue
    pst0 = np.loadtxt(os.path.join(panstarrs_filters, "panStarrs_%s.pb" % ft))
    ax.plot(pst0[:, 0], pst0[:, 1], color=c)
    ax.set_xlim(3000, 11500)
    ax.set_ylim(0, 1)
    ax.set_xlabel("$\lambda$ ($\AA$)", fontsize=16)
    ax.set_ylabel("throughput", fontsize=14)

for fname, c in zip(os.listdir(euclid_filters), ["blue", "green", "red", "orange"]):
    ax = axtup[1][1]
    ax.set_title("Euclid")
    fpath = os.path.join(euclid_filters, fname)
    data = np.loadtxt(fpath)
    ax.plot(data[:, 0], data[:, 1], color=c, label=fname)
    ax.set_xlim(5000, 22000)
    ax.set_ylim(0, 1.05)
    ax.set_xlabel("$\lambda$ ($\AA$)", fontsize=16)
    ax.set_ylabel("throughput", fontsize=14)
    
plt.tight_layout()
plt.savefig("/home/moraes/cosmossnap_filters.png")
```
