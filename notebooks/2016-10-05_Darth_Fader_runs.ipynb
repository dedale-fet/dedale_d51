```python
%matplotlib inline
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
ztrue = fits.getdata("/share/splinter/moraes/Dedale_Project/Darth_Fader/df_data/trueredshifts.fits.gz")
zgen = fits.getdata("/share/splinter/moraes/Dedale_Project/Darth_Fader/df_data/redshifts_gen1.fits")
```

```python
_ = plt.hist(ztrue, bins=np.arange(ztrue.min(), ztrue.max(), 0.01))
plt.figure()
_ = plt.hist(zgen, bins=np.arange(zgen.min(), zgen.max(), 0.01))
```

```python
mask = (np.abs(zgen - ztrue) < 0.002*(1 + ztrue))
plt.plot(ztrue, ztrue, "r--")
plt.scatter(ztrue[mask], zgen[mask], s=2)
plt.xlim(0, 1.8)
plt.ylim(0, 1.8)
plt.figure()
_ = plt.hist(ztrue[mask] - zgen[mask], bins=np.arange(-0.01, 0.01, 0.0005))

print("Percent redshifts with no valid measurement (-1): %.2f" % (100*np.sum(zgen == -1)/len(zgen)))
print("Percent of catastrophic outliers: %.2f" % (100*np.sum(~mask)/len(zgen)))
```
