```python
%matplotlib inline
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.table import Table
import glob
```

```python
flistwide = glob.glob("2017-01-26_TIPS_partial_test_run/Euclid_Wide_Nexp4/clean_TIPSspec_ID_2*")
flistdeep = glob.glob("2017-01-26_TIPS_partial_test_run/Euclid_Deep_Nexp_20/clean_TIPSspec_ID_2*")
```

```python
fig, ((ax1, ax2)) = plt.subplots(nrows=1, ncols=2, figsize=(18, 4))
for fname in flistwide:
    spec = Table.read(fname)
    ax1.plot(spec["wave"], spec["flux"])
    ax2.plot(spec["wave"], spec["error"])
    ax2.set_yscale("log")
    
fig, ((ax1, ax2)) = plt.subplots(nrows=1, ncols=2, figsize=(18, 4))
for fname in flistdeep:
    spec = Table.read(fname)
    ax1.plot(spec["wave"], spec["flux"])
    ax2.plot(spec["wave"], spec["error"])
    ax2.set_yscale("log")
```

```python
flistclean = glob.glob("2017-01-26_euclid_run_individual_spectra_for_TIPS/clean_TIPSspec_ID_2*")[:28]
```

```python
for clean, wide, deep in zip(flistclean[:15], flistwide[:15], flistdeep[:15]):
    plt.figure(figsize=(6, 4))
    spec_wide = Table.read(wide)
    plt.plot(spec_wide["wave"], spec_wide["flux"], "m")
    spec_deep = Table.read(deep)
    plt.plot(spec_deep["wave"], spec_deep["flux"], "y")
    spec_clean = Table.read(clean)
    plt.plot(spec_clean["wave"], spec_clean["flux"], "k")
    plt.yscale("log")
    #plt.xlim(11000, 20000)
    plt.ylim(1e-20, 1e-16)
```
